```python
import bs4
import requests
import pandas
```

```python
soup = bs4.BeautifulSoup(open('india 1 table.html').read(), "html.parser")

soup
```

```python
links = soup.find_all('a', href='javascript:;')
links
```

```python
paginationLinks = []
baseURL = 'http://nabcb.qci.org.in/accreditation/'
for link in links:
    tail = link.get('onclick')[17:link.get('onclick').find('php')+3]
    paginationLinks.append(baseURL + tail)

paginationLinks
```

```python
records = []

for link in paginationLinks:
    soup = bs4.BeautifulSoup(requests.get(link).content)
    
    title = soup.find('td', align='center', class_='heading').get_text()
    tableRows = soup.find_all('td', class_='justify')[-1].find_all('tr')
    for row in tableRows[2:]:
        tableData = row.find_all('td')
        records.append({'Name': title,
                        'IAF Scope': tableData[0].get_text(),
                        'Products / Processes covered in the scope': tableData[1].get_text(),
                        'Description of Scheme': tableData[2].get_text(),
                        'Standards/Regulations/ Normative document(s)': tableData[3].get_text()})

records
```

```python
df = pandas.DataFrame(records)

df.head()
```

```python
df.to_csv('NABCB PCB Excel.csv')
```
