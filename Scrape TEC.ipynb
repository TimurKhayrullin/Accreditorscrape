```python
import bs4
import requests
import pandas
```

```python
soup = bs4.BeautifulSoup(open('india 2 table.html').read(), "html.parser")

soup
```

```python
records = []

tableRows = soup.find_all('tr')

for row in tableRows[1:]:
    tableData = row.find_all('td')
    name = tableData[1].find('strong').get_text()
    address = tableData[1].get_text()[len(name):tableData[1].get_text().find('Tel.:')]
    for ptag in tableData[2].find_all('p'):
        records.append({'Name': name,
                       'Scope': ptag.get_text()})
records
```

```python
df = pandas.DataFrame(records)

df.head()
```

```python
df.to_csv('TEC Excel.csv')
```
